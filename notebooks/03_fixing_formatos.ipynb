```python
import re

def clean_price(value):
    if pd.isna(value):
        return None
    
    value = str(value).strip().lower()
    
    # 1. Convertir palabras a números
    word_map = {
        "twentyfive": 25,
        "twenty five": 25,
        "thirtyfive": 35,
        "thirty five": 35,
        "sixty": 60,
        "sixtyfive": 65,
        "sixty five": 65
    }
    if value in word_map:
        return float(word_map[value])
    
    # 2. Eliminar moneda (sek, €, etc.)
    value = re.sub(r"[a-zA-Z€$]", "", value)
    
    # 3. Convertir a número
    try:
        return float(value)
    except:
        return None
```

```python
bookings_clean["price"] = bookings_clean["price"].apply(clean_price)
```

```python
def clean_discount(value):
    if pd.isna(value) or value == "":
        return 0.0
    
    value = str(value).strip().lower()
    
    # Si viene con %, convertir a decimal
    if "%" in value:
        value = value.replace("%", "")
        return float(value) / 100
    
    return float(value)
```

```python
bookings_clean["discount"] = bookings_clean["discount"].apply(clean_discount)
```

```python
DEFAULT_TAX = 0.25  # o el valor que corresponda

def clean_tax(value):
    if pd.isna(value) or value == "":
        return DEFAULT_TAX
    
    value = str(value).strip().lower()
    
    # Reemplazar coma por punto
    value = value.replace(",", ".")
    
    # Si viene con %, convertir a decimal
    if "%" in value:
        value = value.replace("%", "")
        return float(value) / 100
    
    return float(value)
```

```python
bookings_clean["tax_rate"] = bookings_clean["tax_rate"].apply(clean_tax)
```

```python
def clean_discount(value):
    if pd.isna(value) or value == "":
        return 0.0
    
    value = str(value).strip().lower()
    
    # Reemplazar coma por punto
    value = value.replace(",", ".")
    
    # Si viene con %, convertir a decimal
    if "%" in value:
        value = value.replace("%", "")
        return float(value) / 100
    
    # Si es un número entero (5, 10, 15)
    if value.isdigit():
        return float(value) / 100
    
    # Si ya es decimal (0.1, 0.05)
    try:
        num = float(value)
        # Si es mayor que 1, es porcentaje mal escrito (ej: 5 → 5.0)
        if num > 1:
            return num / 100
        return num
    except:
        return 0.0
```

```python
bookings_clean["discount"] = bookings_clean["discount"].apply(clean_discount)
```

```python
bookings_clean.to_sql(
    "bookings",
    engine,
    if_exists="append",
    index=False
)
```
